```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('ml-1m/users.dat', sep='::', header=None, names=unames, engine='python')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('ml-1m/ratings.dat', sep='::', header=None, names=rnames, engine='python')

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('ml-1m/movies.dat', sep='::', header=None, names=mnames, engine='python')


data = pd.merge(pd.merge(ratings, users), movies)

```

```python
# 5. A function that given a user_id and movie_id , returns a list 
# of user_ids for other users that rated the
# movie identified by the provided movie_id with the same score.
def find_userid_of_similar_ratings(movie_id, user_id, data):
    data_sim = data[(data['movie_id'] == movie_id) & (data['user_id'] != user_id)]
    user_rating = data[(data['movie_id'] == movie_id) & (data['user_id'] == user_id)]['rating']
    ls_user_id = data_sim[data_sim['rating'] == user_rating.iloc[0]]['user_id']
    return ls_user_id
```

```python
# 6.This function is used to plot the graph to indicate the age group of 
# people involved in reviewing the particular genre movie.

# It can be observed from the below graph that the people in age group 20-29 are 
# mostly involved in reviewing most of the gener movies.
def movies_based_on_genre(movie_genre):
  labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
  data['age_group'] = pd.cut(data.age, range(0, 81, 10), right=False, labels=labels)
  count = data.loc[data['genres'] == movie_genre].groupby('age_group').size()
  width = 1/1.5
  plt.bar(labels, count, width, color="blue")

```
