# Principal Component Analysis

Principal component analysis is popular dimensionlity reduction technique. This technique uses an orthogonal transformation to convert observations into linearly uncorrelated variables called principal components.

Here for dimensionality reduction we use PCA instead of thresholding which can be used by clustering algorithms.

```python
# Pandas for DataFrames
import pandas as pd

# NumPy for numerical computing
import numpy as np

# Matplotlib for visualization
import matplotlib.pyplot as plt

# display plots in the notebook
%matplotlib inline

# Seaborn for better visualization of data
import seaborn as sns

# PCA from Scikit-Learn
from sklearn.decomposition import  PCA

# StandardScaler from Scikit-Learn
from sklearn.preprocessing import StandardScaler
warnings.filterwarnings("ignore")
```

```python
# Read product data generated in the previous step 
product_data = pd.read_excel("/Users/ruchitha/Desktop/DataMining_Projects/Online-retail-Analysis/Products_top_30_data.xlsx")
product_data.head()
```

```python
# shape of the dataset considered for PCA
product_data.shape
```

StandardScaler is used to fit and transform product data

```python
# Initialize StandardScaler
sc = StandardScaler()

# Fitting and transforming data
product_data_scaled = sc.fit_transform(product_data)

product_data_scaled = pd.DataFrame(product_data_scaled)

# print first 5 rows
product_data_scaled.head()

product_data_scaled1 = product_data_scaled
```

```python
# Initializing PCA
pca = PCA()

# Generating new set of features
pca_data = pca.fit_transform(product_data_scaled)

pca_data = pd.DataFrame(pca_data)
# print first 5 rows of PCA generated data
pca_data.head()
```

```python
# Cumulative explained variance
cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot cumulative explained variance
plt.plot(range(len(cumulative_explained_variance)), cumulative_explained_variance)

plt.show()
```

```python
# Variance captured for the first 100 components
cumulative_explained_variance[30]
```

```python
# PCA transformation to get 125 components
pca = PCA(n_components=30)

# Fit and transform product data
pca_data1 = pca.fit_transform(product_data_scaled1)

# Shape of PCA transformed data
pca_data1.shape
```

```python
# creating a dataframe of PCA generated components
df_pca_data = pd.DataFrame(pca_data1)

# renaming the columns
df_pca_data.columns = ['PCA{}'.format(i+1) for i in range(df_pca_data.shape[1])]

# fetching indices
df_pca_data.index = product_data.index

# showing first 5 rows
df_pca_data.head(5)
```

```python
# Saving dataset which consists of all the principal features obtained by PCA analysis
# into excel sheet
writer = pd.ExcelWriter('pca_product_data.xlsx')
df_pca_data.to_excel(writer,'Sheet1', index = False)
writer.save()
```
