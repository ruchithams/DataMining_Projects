# Cluster Analysis

In this section we are usign K-means clustering algorithm to classify the given dataset to n clusters. This allows us to classify the datset without having to consider the ground truth. We are considering datasets with different features sets and comparing their results. 

```python
# print_function for compatibility with Python 3
from __future__ import print_function

# numpy for numerical computing
import numpy as np

# importing pandas for dataframe
import pandas as pd

# Matplotlib for visualization
import matplotlib.pyplot as plt

# display plots in the notebook
%matplotlib inline

# Seaborn for visualization
import seaborn as sns

# Scikit-Learn's make_pipeline function
from sklearn.pipeline import make_pipeline

# Scikit-Learn's StandardScaler
from sklearn.preprocessing import StandardScaler

# Scikit-Learn's KMeans algorithm
from sklearn.cluster import KMeans


# Adjusted Rand index
from sklearn.metrics import adjusted_rand_score
warnings.filterwarnings("ignore")
```

```python
# Import customer data 
customer_df = pd.read_excel('/Users/ruchitha/Desktop/DataMining_Projects/Online-retail-Analysis/customer_data.xlsx')

# Import customer data with top 30 products as features
top_30_data =  pd.read_excel('/Users/ruchitha/Desktop/DataMining_Projects/Online-retail-Analysis/Products_top_30_data.xlsx')

# Import PCA item features
pca_data =  pd.read_excel('/Users/ruchitha/Desktop/DataMining_Projects/Online-retail-Analysis/pca_product_data.xlsx')
```

```python
customer_df.head()
```

```python
top_30_data.head()
```

```python
pca_data.head()
```

```python
# Join cutsomer_df with top_30_data
top_30_df = customer_df.join(top_30_data)

# Display first 5 rows of threshold_df
top_30_df.head(5)
```

```python
# Join cutsomer_df with pca_item_data
pca_df = customer_df.join(pca_data)

# Display first 5 rows of pca_df
pca_df.head(5)
```

```python
# K-Means model pipeline
kmeans = make_pipeline(StandardScaler(), KMeans(n_clusters=3, random_state=123))
```

```python
# Fit K-Means pipeline
kmeans.fit(customer_df)
```

```python
# Save clusters to base_df
customer_df['cluster'] = kmeans.predict(customer_df)

# Display first 5 rows of base_df
customer_df.head(5)
```

```python
# Scatterplot, colored by cluster
sns.lmplot(x='total_products',y = 'total_sales', data=customer_df, hue='cluster', fit_reg=False)
plt.show()
```

Clustering appears to seperate customers based on number of items and total amount spent for cutomer dataset which consists of trasaction records

** For PCA generated Features**

```python
# K-Means model pipeline
kmeans = make_pipeline(StandardScaler(), KMeans(n_clusters=3, random_state=123))
```

```python
# Fit K-Means pipeline
kmeans.fit(pca_df)
```

```python
# Save clusters to pca_df
pca_df['cluster'] = kmeans.predict(pca_df)

# Display first 5 rows of pca_df
pca_df.head(5)
```

Clustering appears to seperate customers based on number of items and total amount spent for cutomer dataset which consists of PCA generated principal features

** Model Comparison **

The Rand Index computes a similarity measure between two clusterings by considering all pairs of samples and counting pairs that are assigned in the same or different clusters in the predicted and true clusterings.

```python
adjusted_rand_score(customer_df.cluster, top_30_df.cluster)
```

```python
adjusted_rand_score(top_30_df.cluster, customer_df.cluster)
```

```python
adjusted_rand_score(customer_df.cluster, pca_df.cluster)
```

From the above adjusted random score values we can notice that there is very little similarity between the clusters generated on different dataset samples.
