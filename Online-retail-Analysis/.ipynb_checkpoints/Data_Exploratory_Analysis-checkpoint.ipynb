**Feature-Engineering: Customer-level**

```python
# Grouping data by Customer ID
transaction_data = desired_data.groupby('CustomerID').InvoiceNo.agg({'total_transactions':'nunique'})

# Display transaction data for first 5 customers
transaction_data.head(5)
```

```python
# Fetching total product and unique product count purchased by unique customer
stock_data = desired_data.groupby('CustomerID').StockCode.agg({
                                'total_products':'count',
                                 'total_unique_products':'nunique'})

# Display product data for first 5 customers
stock_data.head(5)
```

```python
# fetching sales data
sales_data = desired_data.groupby('CustomerID').total_cost.agg({
                    'total_sales': 'sum',
                     'avg_product_value': 'mean'
                        })

# Display sales data for first 5 customers
sales_data.head(5)
```

** Feature-Engineering: Cart-Level**

```python
# Group data by cart-level data 
cart_data = desired_data.groupby(['CustomerID', 'InvoiceNo']).total_cost.agg({
                'cart_value': 'sum',
                 } )

# Reset index
cart_data.reset_index(inplace=True)

# Display cart data for first 5 Carts
cart_data.head(5)
```

```python
# Fetching mean, max and min values for cart data 
agg_cart_data = cart_data.groupby('CustomerID').cart_value.agg({
    'avg_cart_value': 'mean',
    'min_cart_value': 'min',
    'max_cart_value': 'max'
})

# Display cart data for first 5 customers
agg_cart_data.head(5)
```

```python
# Join together tables to have consolidated customer data
customer_data = transaction_data.join([stock_data, sales_data, agg_cart_data])

# Display customer-level data for first 5 customers
customer_data.head(5)
```

```python
# writing customer data to excel for future analysis
writer = pd.ExcelWriter('customer_data.xlsx')
customer_data.to_excel(writer,'Sheet1', index = False)
writer.save()
```
